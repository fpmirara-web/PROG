```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file_path = "/content/fbref_PL_2024-25.csv"
df=pd.read_csv(file_path)
```

```python
import plotly.express as px

nation_counts = df["Nation"].value_counts().head(15)

# Dictionary mapping nation codes to flag colors (simplified for demonstration)
flag_colors = {
    "eng ENG": "red",
    "br BRA": "green",
    "fr FRA": "blue",
    "nl NED": "orange",
    "pt POR": "red",
    "dk DEN": "red",
    "es ESP": "red",
    "ie IRL": "green",
    "ar ARG": "lightblue",
    "sct SCO": "blue",
    "de GER": "black",
    "be BEL": "black",
    "gh GHA": "red",
    "se SWE": "blue",
    "wls WAL": "red"
}

# Get colors for the nations in the top 15, default to grey if not in dictionary
colors = [flag_colors.get(nation, "grey") for nation in nation_counts.index]

fig = px.bar(
    x=nation_counts.index,
    y=nation_counts.values,
    color=colors,
    color_discrete_map="identity", # Use the colors provided
    labels={"x": "Nacionalidade", "y": "Número de Jogadores"},
    title="Top 15 Nacionalidades na Premier League 2024/25"
)

fig.update_layout(xaxis_tickangle=45)
fig.show()
```
