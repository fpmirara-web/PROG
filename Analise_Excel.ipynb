```python
import pandas as pd
```

```python
df=pd.read_excel('/content/Chamados ficticios - Cora (Dez_2020).xlsx')
```

```python
df
```

```python
df=df.drop(columns=["SequentialId","Team" ,"Closed"])
```

```python
df
```

```python
df=df.dropna()
```

```python
df
```

```python
df.info()
```

```python
df['AverageResponseTime'].apply(lambda x: x [0:-2]).astype(float)
```

```python
df
```

```python
df['AverageResponseTime'] /=60
```

```python
df
```

```python
df_tempo_por_agente = df.groupby("AgentName")["AverageResponseTime"].mean()
```

```python
df_tempo_por_agente
```

```python
df_tempo_por_agente = df_tempo_por_agente.sort_values('AverageResponseTime')
```

```python
df_tempo_por_agente
```

```python
df_tempo_por_agente.plot(kind="barh",figsize=(12,6))
```
