```python
import pandas as pd
import geopandas as gpd
import folium
```

```python
df=pd.read_csv(r"/content/volcano_data_2010.csv")
```

```python
df
```

```python
df= df [["Year", "Name", "Country", "Latitude", "Longitude"]]
```

```python
df
```

```python
geometrias=gpd.points_from_xy(df['Latitude'], df['Longitude'])
```

```python
geo_df=gpd.GeoDataFrame(df, geometry=geometrias)
geo_df
```

```python
Mapa=folium.Map(location= [15, 30], tiles="Cartodb dark_matter", zoom_start=2)
heatmap_dados= [[point.xy[1][0], point.xy[0][0]] for point in geo_df["geometry"]]
folium.plugins.HeatMap(heatmap_dados).add_to(Mapa)

```
