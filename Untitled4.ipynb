```python
import pandas as pd
```

```python
df=pd.read_csv('brasileirao.csv')
```

```python
df
```

```python
champions = df.loc[df.groupby('season')['points'].idxmax()]
display(champions[['season', 'team', 'points']].reset_index(drop=True))
```

```python
most_titles = champions['team'].value_counts().sort_values(ascending=False)
display(most_titles)
```

```python
df['win_percentage'] = (df['won'] / df['played']) * 100
best_teams_percentage = df.sort_values(by='win_percentage', ascending=False)
display(best_teams_percentage[['team', 'season', 'win_percentage']].reset_index(drop=True))
```

```python
# Assuming the bottom 4 teams are relegated in each season
relegated_teams = df.groupby('season').tail(4)

worst_campaigns = relegated_teams.sort_values(by='points', ascending=True)
display(worst_campaigns[['season', 'team', 'points', 'place']].reset_index(drop=True))
```

```python
champions['avg_goals_per_game'] = champions['goals'] / champions['played']
best_avg_goals_champion = champions.loc[champions['avg_goals_per_game'].idxmax()]
display(best_avg_goals_champion[['season', 'team', 'avg_goals_per_game']])
```

```python
best_defense_champion = champions.loc[champions['goals_taken'].idxmin()]
display(best_defense_champion[['season', 'team', 'goals_taken']])
```

```python
champions['win_percentage'] = (champions['won'] / champions['played']) * 100
display(champions[['season', 'team', 'win_percentage']].reset_index(drop=True))
```

```python
worst_champion_campaign = champions.loc[champions['points'].idxmin()]
display(worst_champion_campaign[['season', 'team', 'points']])
```

```python
import plotly.express as px

champions_sorted_by_points = champions.sort_values(by='points', ascending=True)

fig = px.bar(champions_sorted_by_points,
             x=champions_sorted_by_points['season'].astype(str) + ' - ' + champions_sorted_by_points['team'],
             y='points',
             title='Points of Brasileirão Champions (Worst to Best Campaign)',
             labels={'x': 'Season - Team', 'y': 'Points'},
             hover_data={'points': True}) # Show points on hover

fig.update_layout(xaxis={'categoryorder':'total ascending'}) # Ensure bars are ordered correctly
fig.show()
```

```python
best_single_season_wins = df.loc[df['won'].idxmax()]
display(best_single_season_wins[['season', 'team', 'won']])
```

```python
worst_champion_losses = champions.loc[champions['loss'].idxmax()]
display(worst_champion_losses[['season', 'team', 'loss']])
```

```python
worst_champion_defense = champions.loc[champions['goals_taken'].idxmax()]
display(worst_champion_defense[['season', 'team', 'goals_taken']])
```

# Task
Transfer the results of the analysis on Brazilian football championship data, including champions, most titles, worst campaigns, best attacks, best defenses, etc., into a single Excel file with different sheets for each analysis result.

## Consolidate analysis results

### Subtask:
Gather the relevant DataFrames and Series containing the results of the analysis (champions, most titles, worst campaigns, best attacks, best defenses, etc.).


**Reasoning**:
Create a dictionary to store the analysis results for writing to an Excel file.



```python
analysis_results = {
    'Champions': champions[['season', 'team', 'points', 'win_percentage']],
    'Most Titles': most_titles.rename('Number of Titles'),
    'Worst Campaigns': worst_campaigns[['season', 'team', 'points', 'place']],
    'Best Win Percentage (All Seasons)': best_teams_percentage[['team', 'season', 'win_percentage']],
    'Best Champion Attack (Avg Goals per Game)': best_avg_goals_champion.to_frame().T[['season', 'team', 'avg_goals_per_game']],
    'Best Champion Defense (Goals Taken)': best_defense_champion.to_frame().T[['season', 'team', 'goals_taken']],
    'Worst Champion Campaign (Points)': worst_champion_campaign.to_frame().T[['season', 'team', 'points']],
    'Best Single Season Wins': best_single_season_wins.to_frame().T[['season', 'team', 'won']],
    'Worst Champion Losses': worst_champion_losses.to_frame().T[['season', 'team', 'loss']],
    'Worst Champion Defense (Goals Taken)': worst_champion_defense.to_frame().T[['season', 'team', 'goals_taken']]
}
```

## Create an excel writer object

### Subtask:
Initialize a pandas Excel writer to save multiple sheets to a single Excel file.


**Reasoning**:
Initialize a pandas Excel writer to save multiple sheets to a single Excel file.



```python
from pandas.io.excel import ExcelWriter

excel_file = ExcelWriter('brasileirao_analysis.xlsx')
```

## Write results to different sheets

### Subtask:
Write each set of results to a separate sheet in the Excel file.


**Reasoning**:
Iterate through the analysis results dictionary and write each result to a separate sheet in the Excel file, handling the index based on the specific result.



```python
for sheet_name, data in analysis_results.items():
    if sheet_name == 'Most Titles':
        data.to_excel(excel_file, sheet_name=sheet_name, index=True)
    else:
        data.to_excel(excel_file, sheet_name=sheet_name, index=False)

excel_file.save()
```

**Reasoning**:
The traceback indicates that the `save()` method is not available for the `OpenpyxlWriter` object. The correct method to use with newer versions of pandas and openpyxl is `close()`. The UserWarning about the sheet title length is also noted, but the primary error is the `AttributeError`. I will replace `save()` with `close()` and keep the long sheet names for now, as it's just a warning and the main goal is to write the data.



```python
for sheet_name, data in analysis_results.items():
    if sheet_name == 'Most Titles':
        data.to_excel(excel_file, sheet_name=sheet_name, index=True)
    else:
        data.to_excel(excel_file, sheet_name=sheet_name, index=False)

excel_file.close()
```
